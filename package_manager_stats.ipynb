# Package Manager Stats for Nexmo DevRel Packages

Data from https://nexmo-devrel-stats.herokuapp.com/stats

Charts for: [NodeJS](#Node-Libraries) - [PHP](#PHP-Libraries) - [Python](#Python-Libraries) - [Ruby](#Ruby-Libraries) (so far)

```python
import matplotlib.pyplot as plt
import pandas as pd
package_stats = pd.read_json('https://nexmo-devrel-stats.herokuapp.com/stats')
package_stats.tail()
package_stats['created'] = pd.to_datetime(package_stats['created'], dayfirst="true")
#package_stats.info()
#package_stats.tail()
```

## PHP Libraries

```python
is_php = package_stats['language']=="php"
php_packages = package_stats[is_php]
```

```python
df = php_packages.pivot(index="created", columns="library", values="month_total")
df.plot(rot=90)
plt.ylim(bottom=0)
```

## Node Libraries

```python
is_node = package_stats['language']=="node"
node_packages = package_stats[is_node]
```

```python
df = node_packages.pivot(index="created", columns="library", values="month_total")
df.plot(rot=90)
plt.ylim(bottom=0)
```
